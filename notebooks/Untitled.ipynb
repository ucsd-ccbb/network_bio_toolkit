```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C) # just plot some lines straight from the data
plt.plot(X, S)

plt.show()
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
sub = plt.subplot(1, 1, 1) # guess just do this...

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # data to plot
C, S = np.cos(X), np.sin(X)

# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-") # plot that data

# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Set x limits
plt.xlim(-4.0, 4.0)

# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True)) # min_range, max_range, # of ticks

# Set y limits
plt.ylim(-1.0, 1.0) # actual size cut-off

# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# Save figure using 72 dots per inch
# plt.savefig("exercice_2.png", dpi=72)

# Show result on screen
plt.show()
```

```python
sns.distplot(C, kde = True, ax = sub)
plt.show()
```
